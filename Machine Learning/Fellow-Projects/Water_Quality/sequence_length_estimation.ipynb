Following the guidance of Dr. Samadi, in this script we calculate Time of Concentration for our watershed and use it as the value for the Sequence Length Parameter.

We are using the Kirpich Equation from Chapter 15 of the National Engineering Handbook (https://irrigationtoolbox.com/NEH/Part630_Hydrology/NEH630-ch15draft.pdf)

We estimated *L*, the length of the flowpath from the headwater to the catchment outlet in feet, and *S*, the slope of the longest hydraulic length in ft/ft, using ArcGIS Pro and a 10-m DEM from the USGS.

Kirpich_Screenshot.png

```python
L = 1245964.881 # feet
S = 22.171832 / 100 # percent converted to ft/ft
Tc = 0.0078 * L**(0.77) * S**(-0.385) # minutes
Tc_days = Tc / 1440 # days

print(round(Tc_days,2), "days")
```
