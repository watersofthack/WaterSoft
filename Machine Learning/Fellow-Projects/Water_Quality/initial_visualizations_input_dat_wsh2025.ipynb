```python
# read data
dat_daily = pd.read_csv('11447650_cleaned_daily_data_plus_precipitation.csv',
                        parse_dates = ['DATE'])
dat_iv = pd.read_csv('11447650_raw_iv_turb.csv',
                     parse_dates = ['datetime'])
```

```python
# plot precip
plt.figure(figsize = (8, 4))
plt.plot(dat_daily['DATE'], dat_daily['Precip_in'])
plt.xlabel('Date')
plt.ylabel('Precipitation (in)')
plt.title('NOAA Precipitation Record')
plt.grid()
plt.show()

# plot ssd
plt.figure(figsize = (8, 4))
plt.plot(dat_daily['DATE'], dat_daily['SSD_st_d'])
plt.xlabel('Date')
plt.ylabel('SSD')
plt.title('USGS Suspended Sediment Discharge')
plt.grid()
plt.show()

# plot SSC_mg_L
plt.figure(figsize = (8, 4))
plt.plot(dat_daily['DATE'], dat_daily['SSC_mg_L'])
plt.xlabel('Date')
plt.ylabel('SSC (mg/L)')
plt.title('USGS Suspended Sediment Concentration')
plt.grid()
plt.show()

# plot discharge
plt.figure(figsize = (8, 4))
plt.plot(dat_daily['DATE'], dat_daily['Discharge_cfs'])
plt.xlabel('Date')
plt.ylabel('Q (cfs)')
plt.title('USGS Discharge')
plt.grid()
plt.show()

# plot turb
plt.figure(figsize = (8, 4))
plt.plot(dat_iv['datetime'], dat_iv['turbidity'])
plt.xlabel('Date')
plt.ylabel('Turbidity')
plt.title('USGS Instantaneous Turbidity')
plt.grid()
plt.show()
```
